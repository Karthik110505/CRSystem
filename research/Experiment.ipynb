```python
def recommend(college_name):
    college_index = np.where(ds_schools==college_name)[0][0]
    college_list  = sorted(list(enumerate(distances[college_index])),reverse=True,key=lambda x : x[1])[0:6]
    college_list = [i[0] for i in college_list]
    print(college_list)
```

```python
ds_original["Need_GRE"]=ds_original["Need_GRE"].apply(lambda x : "Needed" if x ==1 else "Not Needed")
ds_original["Institution Type"]=ds_original["Institution Type"].apply(lambda x : "Public" if x ==1 else "Private")
ds_original["Online"]=ds_original["Online"].apply(lambda x : "Online" if x ==1 else "Offline")
ds_original.head(1)
```

```python
from urllib.parse import urlparse

def extract_base_url(url):
    parsed_url = urlparse(url)
    base_url = parsed_url.scheme + '://' + parsed_url.netloc
    return base_url

small['LINK'] = small['LINK'].apply(extract_base_url)

```

```python
link = small[["SCHOOL","LINK"]]
link=link.drop_duplicates(subset='SCHOOL')
ds_original=pd.merge(ds_original,link,left_on ="School Name",right_on="SCHOOL",how="left").drop(["SCHOOL"],axis="columns").fillna("NA")
```

```python
csv_file_name = 'data.csv'
ds_original.to_csv(csv_file_name, index=False)
```
