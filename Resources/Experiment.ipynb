```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_target = label_encoder.fit_transform(dsprogram['School Name'])
encoded_target
```

```python
dsprogram.drop(["School Name"],axis="columns",inplace=True)
```

```python
ds_encoded=pd.get_dummies(dsprogram)
```

```python
ds_encoded
```

```python
boolean_columns = [col for col in ds_encoded.columns if ds_encoded[col].dtype == bool]
ds_encoded[boolean_columns] = ds_encoded[boolean_columns].astype(int)
ds_encoded
```

```python
ds_encoded.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds_encoded[["Ranking","Total_Tuition_Cost","Median_Salary_10yr"]]  =sc.fit_transform(ds_encoded[["Ranking","Total_Tuition_Cost","Median_Salary_10yr"]])
```

```python
ds_encoded
```

```python
from sklearn.ensemble import RandomForestClassifier
rand =RandomForestClassifier()
rand.fit(ds_encoded.drop(["Ranking"],axis="columns"),encoded_target)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(rand.predict(ds_encoded.drop(["Ranking"],axis="columns")),encoded_target))
```
